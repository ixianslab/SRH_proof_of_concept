## Aufgabe : simulierter Würfelwurf
Generieren Sie in Python 1000 (gleichwahrscheinliche) Würfe <br> a) eines Würfels und berechnen Sie Mittelwert und Standardabweichung der Würfelsumme. Wählen Sie aus den Würfelwürfen 200 Stichproben vom Umfang 50 aus und berechnen deren Mittelwert und Standardfehler. Vergleichen Sie die ursprüngliche Verteilung graphisch mit dem Histogramm der Verteilung der Stichprobenmittelwerte. <br> Wiederholen Sie das Experiment für <br> b) 10 Würfel <br>  c) eine Münze. 

```python
import random
import numpy as np
import matplotlib.pyplot as plt



wuerfel1 = []
wuerfel10 = []
muenze = []
#m = input ( 'Anzahl der Würfelseiten :' )

wuerfel1_sample = []
wuerfel10_sample = []
muenze_sample = []
wuerfel1_mean_sample = []
wuerfel10_mean_sample = []
muenze_mean_sample = []
##################################################
for i in range(1000):
    wuerfel1.append(random.randint(1,6))
    wuerfel10.append(random.randint(10,60))
    muenze.append(random.randint(1,2))
    
##################################################   
#print("1000 Würfelwurf mit 1 Würfeln" , wuerfel1)
#print("1000 Würfelwurf mit 10 Würfeln" , wuerfel10)
#print('1000 Würfe einer Münze' , muenze)
##################################################
print('-------------------------------------------------------------')
mean_wuerfel1 = np.mean(wuerfel1)
print("Mittelwert 1 Würfel :" , mean_wuerfel1)
dev_wuerfel1 = np.std(wuerfel1)
print("Standardabweichung 1 Würfel:" , dev_wuerfel1)
print('-------------------------------------------------------------')
##################################################
print('-------------------------------------------------------------')
mean_wuerfel10 = np.mean(wuerfel10)
print("Mittelwert von 10 Würfeln :" , mean_wuerfel10)
dev_wuerfel10 = np.std(wuerfel10)
print("Standardabweichung 10 Würfel:" , dev_wuerfel10)
print('-------------------------------------------------------------')
##################################################
print('-------------------------------------------------------------')
mean_muenze = np.mean(muenze)
print("Mittelwert von Münzwurf :" , mean_muenze)
dev_muenze = np.std(muenze)
print("Standardabweichung Münze:" , dev_muenze)
print('-------------------------------------------------------------')

for i in range(1,201):
    wuerfel1_sample = np.random.choice(wuerfel1 , size=50)
    wuerfel1_mean = np.mean(wuerfel1_sample)
    wuerfel1_mean_sample.append(wuerfel1_mean)
    
    wuerfel10_sample = np.random.choice(wuerfel10 , size=50)
    wuerfel10_mean = np.mean(wuerfel10_sample)
    wuerfel10_mean_sample.append(wuerfel10_mean)
    
    muenze_sample = np.random.choice(muenze , size=50)
    muenze_mean = np.mean(muenze_sample)
    muenze_mean_sample.append(muenze_mean)
print('-------------------------------------------------------------')
print('Plot der Verteilung von 200 Würfen eines Würfels')
plt.plot(wuerfel1)
plt.show()
print('Plot der Verteilung der Mittelwerte einer Stichprobe vom Umfang n = 200 eines Würfels')
plt.hist(wuerfel1_mean_sample, bins = 100)
plt.xlim(3,4)
plt.show()

print('-------------------------------------------------------------')
print('Plot der Verteilung von 200 Würfen von 10 Würfeln')
plt.plot(wuerfel10)
plt.show()
print('Plot der Verteilung der Mittelwerte einer Stichprobe vom Umfang n = 200 von 10 Würfeln')
plt.hist(wuerfel10_mean_sample, bins = 100)
plt.xlim(30,40)
plt.show()
print('-------------------------------------------------------------')
print('Plot der Verteilung von 200 Würfen einer Münze')
plt.plot(muenze)
plt.show()
print('Plot der Verteilung der Mittelwerte einer Stichprobe vom Umfang n = 200 einer Münze')
plt.hist(muenze_mean_sample, bins = 100)
plt.xlim(1.2,1.8)
plt.show()
```
