## Aufgabe Konfidenzintervall : $\mu \pm \sigma$
Berechnen Sie für die Grundgesamtheit der Menge {$10, 20, \dots, 990 , 1000$} Mittelwert $\mu$ und Standardabweichung $\sigma$. Nehmen Sie 100 zufällige Stichproben von 5 Elementen und berechnen deren Mittelwert $\mu_{\bar{x}}$. <br> Wieviele Prozent der Stichprobenmittelwerte liegen im Intervall $\mu \pm \sigma_{\bar{x}}$ ?

## Lösung :

```python
import numpy as np

a=list(range(10,1010,10))
b=[]
mean_pop= np.mean(a)
std_pop= np.std(a)
std_sample = std_pop/(5)**(1/2)
low = mean_pop - std_sample
high = mean_pop + std_sample
print(a)
#print(low , 'low')
#print(high , 'high')
print(mean_pop, ': Mittelwert der Grundgesamtheit')
print(std_pop, ': Standardabweichung der Grundgesamtheit')
print(std_sample, ': Standardabweichung des Stichprobenmittels')



for i in range(1,101):
    my_sample_10 = np.random.choice(a , size=5)
    print(my_sample_10,'=',np.mean(my_sample_10))
    if np.mean(my_sample_10) < low :
        print('too low')
    elif np.mean(my_sample_10) > high :
        print('too high')
    else : b.append(np.mean(my_sample_10))
print(b)
#Prozent der Werte innerhalb des Intervalls
frequency=len(b)/100
print(frequency*100 , 'Prozent der Werte liegen innerhalb einer Standardabweichung' )
```
